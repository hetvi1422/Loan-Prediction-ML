### Feature Enginnering

```python
df1['total_assets'] = (
    df1[' residential_assets_value'] +
    df1[' commercial_assets_value'] +
    df1[' luxury_assets_value'] +
    df1[' bank_asset_value']
)
```

```python
df1['loan_income_ratio'] = df1[' loan_amount'] / df1[' income_annum']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df1['education'] = le.fit_transform(df1[' education'])
df1['self_employed'] = le.fit_transform(df1[' self_employed'])
df1['loan_status'] = le.fit_transform(df1[' loan_status'])
```

```python
df1.drop('loan_id', axis=1, inplace=True)
```

```python
df1['income_per_dependent'] = df1[' income_annum'] / (df1[' no_of_dependents'] + 1)
```
